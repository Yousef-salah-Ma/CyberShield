```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('/content/csv/data.csv')
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

df['hour'] = df['Timestamp'].dt.hour
df['day'] = df['Timestamp'].dt.day
df['weekday'] = df['Timestamp'].dt.weekday
```

```python
drop_col = ["Flow ID" , "Src IP"  ,"Dst IP" , "Timestamp"]
df =df.drop(drop_col , axis=1)
```

```python
cat_col = df.select_dtypes(include="object").columns
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in cat_col:
    df[col] = le.fit_transform(df[col])
```

```python
y_label = df['Label']
X = df.drop(['Label',"Traffic Type" , "Traffic Subtype"], axis=1)
```

```python
y_Traffic_Type = df['Traffic Type']
y_Traffic_subtype = df['Traffic Subtype' ]
```

```python
from sklearn.preprocessing import StandardScaler

# scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
y_label.value_counts()
```

```python
import keras

model = keras.models.load_model("my_model.keras")
```

```python
y_pred_label_, y_pred_type_, y_pred_subtype_ = model.predict(X_scaled)
```

```python
import numpy as np

# label binary
y_pred_label_class_ = (y_pred_label_ > 0.6).astype(int)

# Traffic Type
y_pred_type_class_= np.argmax(y_pred_type_, axis=1)

# Traffic Subtype
y_pred_subtype_class_ = np.argmax(y_pred_subtype_, axis=1)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# label
print(confusion_matrix(y_label, y_pred_label_class_))
print(classification_report(y_label, y_pred_label_class_))

# Traffic Type
print(confusion_matrix(y_Traffic_Type, y_pred_type_class_))
print(classification_report(y_Traffic_Type, y_pred_type_class_))

# Traffic Subtype
print(confusion_matrix(y_Traffic_subtype, y_pred_subtype_class_))
print(classification_report(y_Traffic_subtype, y_pred_subtype_class_))
```
